```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ['msft', 'orcl', 'nvda', 'gct', 'aapl']
df = yf.download(tickers=tickers, start='2000-01-01', end='2023-01-01')
```

```python
df.Close
```

```python
all_data = pd.DataFrame(df)
all_data
```

```python
data = all_data.Close.copy()
data
```

```python
aapl = data.AAPL.to_frame()
aapl
```

```python
# percentage change with lag
actual = aapl.AAPL
diff_next_day = aapl.AAPL.shift(1)
aapl['change'] = ((actual - diff_next_day)/diff_next_day)*100
aapl
```
