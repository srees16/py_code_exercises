### Cumulative returns & drawdowns

```python
aapl = yf.download('AAPL')
```

```python
aapl = aapl['Close']['AAPL'].to_frame()
```

```python
aapl.columns = ['Close']
aapl
```

```python
aapl['D_Returns'] = np.log(aapl['Close'].div(aapl['Close'].shift(1)))
aapl.dropna(inplace=True)
aapl
```

```python
# what is the cumulative returns for $1 invested in beginning?
np.exp(aapl.D_Returns.sum()), aapl.D_Returns.sum()
```

```python
aapl['cumulative_returns'] = aapl.D_Returns.cumsum().apply(np.exp)
aapl
```

```python
aapl['cumulative_returns'].plot(figsize=(15, 8), title='If you buy & hold from day 1')
plt.show()
```

```python
# annual mean return and std deviation
aapl['D_Returns'].mean()*252, aapl['D_Returns'].std()*np.sqrt(252)
```

## draw down

```python
aapl['cum_max'] = aapl['cumulative_returns'].cummax()
aapl
```

```python
aapl[['cumulative_returns', 'cum_max']].plot(figsize=(15, 8), title='Cumulative max vs cumulative returns')
plt.show()
```

```python
aapl['draw_down'] = aapl['cum_max'] - aapl['cumulative_returns']
aapl
```

```python
# what is the maximum drawdown?
aapl.draw_down.max()
```

```python
# when did maximum drawdown happen?
aapl.draw_down.idxmax()
```

```python
aapl.loc[aapl.index==aapl.draw_down.idxmax()]
```

```python
# draw down in % wise
aapl['draw_down_%'] = (aapl['cum_max'] - aapl['cumulative_returns'])/aapl['cum_max']
aapl
```

```python
aapl['draw_down_%'].max()
```

```python
aapl.loc[aapl.index==aapl['draw_down_%'].idxmax()]
```

## Challenge 3
## Find best SMA for top 5 stocks
## create a bias strategy
## Compare drawdows
## pick top 3 stocks for portfolio

```python
import yfinance as yf
```

```python
# function
def test_strategy(stock, start_date, end_date, sma):
    data = yf.download(stock, start=start_date, end=end_date)
    #data = df['Close']
    data['returns'] = np.log(data.Close.div(data.Close.shift(1)))
    data['sma_s'] = data['Close'].rolling(int(sma[0])).mean()
    data['sma_l'] = data['Close'].rolling(int(sma[1])).mean()
    data.dropna(inplace=True)
    data['position'] = np.where(data['sma_s'] > data['sma_l'], 1, -1)
    data['strategy'] = data['returns'] * data.position.shift(1)
    data.dropna(inplace=True)
    returns = np.exp(data['strategy'].sum())
    std_dev = data['strategy'].std()*np.sqrt(252)
    return returns, std_dev
```

```python
test_strategy('nvda', '2010-01-01', '2025-01-31', (50, 200))
```

## Class

```python
class SMABackTester():

    def __init__(self, symbol, sma_s, sma_l, start, end):
        self.symbol = symbol
        self.sma_s = sma_s
        self.sma_l = sma_l
        self.start = start
        self.end = end
        self.results = None
        self.get_data()

    def get_data(self):
        data = yf.download(self.symbol, start = self.start, end = self.end)
        #data = df['Close']
        data['returns'] = np.log(data['Close'].div(data['Close'].shift(1)))
        data['sma_s'] = data['Close'].rolling(self.sma_s).mean()
        data['sma_l'] = data['Close'].rolling(self.sma_l).mean()
        data.dropna(inplace=True)
        self.data2 = data
        return data

    def test_results(self):
        data = self.data2.copy().dropna()
        data['position'] = np.where(data['sma_s'] > data['sma_l'], 1, -1)
        data['strategy'] = data['returns'] * data.position.shift(1)
        data.dropna(inplace=True)
        data['returnsbh'] = data['returns'].cumsum().apply(np.exp)
        data['returns_strategy'] = data['strategy'].cumsum().apply(np.exp)
        performance = data['returns_strategy'].iloc[-1]
        outperformance = performance - data['returnsbh'].iloc[-1]
        self.results = data
        returns = np.exp(data['strategy'].sum())
        std_dev = data['strategy'].std()*np.sqrt(252)
        return round(performance, 6), round(outperformance, 6)

    def plot_results(self):
        if self.results is None:
            print('Please run the test')
        else:
            title = '{} | SMA_S = {} | SMA_L = {}'.format(self.symbol, self.sma_s, self.sma_l)
            self.results[['returnsbh', 'returns_strategy']].plot(title=title, figsize=(12,8))
            plt.show()
```

```python
tester = SMABackTester('SPY', 50, 100, '2000-01-01', '2020-01-01')
```

```python
tester.get_data()
```

```python
tester.test_results()
```

```python
tester.plot_results()
```
