```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 8,6
import seaborn as sb
sb.set()
```

```python
gld = pd.DataFrame(yf.download('GLD', start='2016-07-01', end='2021-06-30')['Close'])
```

```python
gld.head()
```

```python
gld.columns = ['Close']
gld
```

```python
sma = 21
gld['returns'] = np.log(gld["Close"]).diff()
gld['sma'] = gld['Close'].rolling(sma).mean()
gld['ratio'] = gld['Close'] / gld['sma']
gld
```

```python
gld.describe()
```

```python
gld['ratio'].describe()
```

```python
percentiles = [5, 10, 50, 90, 95]
p = np.percentile(gld['ratio'].dropna(), percentiles)
print(p)
```

```python
gld['ratio'].dropna().plot(legend = True)
plt.axhline(p[0], c= (.5,.5,.5), ls='--')
plt.axhline(p[2], c= (.5,.5,.5), ls='--')
plt.axhline(p[-1], c= (.5,.5,.5), ls='--')
```

```python
short = p[-1]
long = p[0]
gld['position'] = np.where(gld.ratio > short, -1, np.nan)
gld['position'] = np.where(gld.ratio < long, 1, gld['position'])
gld['position'] = gld['position'].ffill()
```

```python
gld.position.dropna().plot()
```

```python
gld['MR_strategy_return'] = gld['returns'] * gld['position'].shift()
```

```python
plt.plot(np.exp(gld['returns'].dropna()).cumprod(), label='Buy/Hold')
plt.plot(np.exp(gld['MR_strategy_return'].dropna()).cumprod(), label='MR Strategy')
plt.legend()
```

```python
print(np.exp(gld['returns'].dropna()).cumprod()[-1] -1)
print(np.exp(gld['MR_strategy_return'].dropna()).cumprod()[-1] -1)
```
